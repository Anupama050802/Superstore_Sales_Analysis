About Dataset(From Kaggle)

"Superstore Sales" generally refers to the sales data and performance of a superstore or a large retail store that offers a wide range of products and merchandise. These superstores are typically known for their extensive inventory, competitive pricing, and one-stop shopping experience.

Analyzing superstore sales data is essential for store owners, managers, and analysts to understand the store's performance, identify trends, and make informed business decisions. Here are some key aspects of a US based superstore sales analysis.

- Row ID => Unique ID for each row.
- Order ID => Unique Order ID for each Customer.
- Order Date => Order Date of the product.
- Ship Date => Shipping Date of the Product.
- Ship Mode=> Shipping Mode specified by the Customer.
- Customer ID => Unique ID to identify each Customer.
- Customer Name => Name of the Customer.
- Segment => The segment where the Customer belongs.
- Country => Country of residence of the Customer.
- City => City of residence of of the Customer.
- State => State of residence of the Customer.
- Postal Code => Postal Code of every Customer.
- Region => Region where the Customer belong.
- Product ID => Unique ID of the Product.
- Category => Category of the product ordered.
- Sub-Category => Sub-Category of the product ordered.
- Product Name => Name of the Product
- Sales => Sales of the Product.
- Quantity => Quantity of the Product.
- Discount => Discount provided.
- Profit => Profit/Loss incurred.

# Business Problem 1: Overall Business Performance

**1. Total Sales Revenue**

```python
q1="""SELECT ROUND(SUM(Sales),2) AS Total_Sales
FROM orders;"""
df1 = pd.read_sql_query(q1,conn)
df1
```

Business Insight

The superstore generated $2297200.86 in total sales over the period 2014 to 2017 covered by the dataset.

**2. Total Profit**

```python
q2="""SELECT ROUND(SUM(Profit),2) AS Total_Profit
FROM orders;"""
df2=pd.read_sql_query(q2,conn)
df2
```

Business Insight

The superstore earned a total profit of 286,397.02 during the analysis period. This indicates that while total sales were over $2.26 million, only about 12.66% of sales translated into profit.

**3. How many orders were placed?**

```python
q3="""SELECT COUNT(DISTINCT "Order ID") AS Total_Orders
FROM orders;"""
df3=pd.read_sql_query(q3,conn)
df3
```

Business Insight

The superstore processed 5,009 unique customer orders

**4. How many unique customers purchased?**

```python
q4="""SELECT COUNT(DISTINCT "Customer ID") AS Total_Customers
FROM orders;"""
df4=pd.read_sql_query(q4,conn)
df4
```

Business Insight

A total of 793 unique customers purchased from the superstore during those 3 years.

**5. Average order value**

```python
q5="""SELECT
ROUND(SUM(Sales)/COUNT(DISTINCT "Order ID"),2) AS Avg_Order_Value
FROM orders;"""
df5=pd.read_sql_query(q5,conn)
df5
```

Business Insight

The average order value (AOV) of $458.61 indicates the typical revenue generated per order. This metric helps evaluate customer purchasing behavior and can be used to measure the effectiveness of pricing, promotions, and cross-selling strategies.

# Business Problem 2: Product Performance

**6. Which product categories generate the highest sales?**

```python
q6="""SELECT
Category,
ROUND(SUM(Sales),2) AS Sales
FROM orders
GROUP BY Category
ORDER BY Sales DESC;"""
df6=pd.read_sql_query(q6,conn)
df6
```

Business Insight

This indicates that Technology products were the primary driver of the Superstore's sales revenue during the analysis period.

**7. Which categories generate the highest profit?**

```python
q7="""SELECT
Category,
ROUND(SUM(Profit),2) AS Profit
FROM orders
GROUP BY Category
ORDER BY Profit DESC;"""
df7=pd.read_sql_query(q7,conn)
df7
```

Business Insight

Technology has generated the highest profit being the category with highest sales but the major observation here is that even though the other cateogory furniture had higher sales than office supplies it generated a lesser profit than office supplies in the analysis period. This suggests that Furniture may have lower profit margins, higher discounts, or higher operational costs.

**8. Which sub-categories perform best?**

```python
q8="""SELECT
"Sub-Category",
ROUND(SUM(Sales),2) AS Sales
FROM orders
GROUP BY "Sub-Category"
ORDER BY Sales DESC;"""
df8=pd.read_sql_query(q8,conn)
df8
```

Business Insight

Phones was the highest-selling sub-category, generating 330,007.05 in sales, closely followed by Chairs with $328,449.10. These two sub-categories were the primary revenue drivers for the business.

Storage, Tables, and Binders also contributed significantly, each generating over $200,000 in sales.

On the other hand, Fasteners, Labels, and Envelopes recorded the lowest sales, indicating relatively low customer demand compared to other product groups.

**9. Which products generate the most revenue?**

```python
q9=""" SELECT "Product Name", ROUND(SUM(Sales),2) AS Sales
FROM Orders
GROUP BY "Product Name"
ORDER BY Sales DESC
LIMIT 10"""
df9=pd.read_sql_query(q9,conn)
df9
```

Business Insight

The Canon imageCLASS 2200 Advanced Copier was the highest-selling product, generating $61,599.82 in sales, making it the store's top revenue contributor.

Office equipment and technology products dominate the top-selling products list, with copiers, printers, video conferencing systems, and binding machines contributing significantly to overall revenue.

The presence of multiple GBC, HP, Canon, and Cisco products among the top performers indicates strong customer demand for high-value office technology and business equipment.

**10. Which products are causing losses?**

```python
q10="""SELECT
"Product Name",
ROUND(SUM(Profit),2) AS Profit
FROM orders
GROUP BY "Product Name"
ORDER BY Profit
LIMIT 10;"""
df10=pd.read_sql_query(q10,conn)
df10
```

Business Insight

The Cubify CubeX 3D Printer Double Head Print recorded the highest loss, with a total profit of -$8,879.97, making it the least profitable product in the dataset.

Several high-value technology products and office furniture appear among the top loss-making products, indicating that expensive products do not always generate positive profits.

These losses may be attributed to heavy discounts, high procurement costs, low profit margins, or frequent returns. Further analysis is needed to determine the exact cause.

# Advanced

**20. Rank categories by sales**

```python
q20="""SELECT
Category,
SUM(Sales) AS Sales,
RANK() OVER(ORDER BY SUM(Sales) DESC) AS Rank
FROM orders
GROUP BY Category;"""
df20=pd.read_sql_query(q20,conn)
df20
```

**21. Customers whose sales are above the average customer sales**

```python
q21="""SELECT
"Customer Name",
SUM(Sales) AS Sales
FROM orders
GROUP BY "Customer Name"
HAVING SUM(Sales) >
(
SELECT AVG(CustomerSales)
FROM
(
SELECT SUM(Sales) AS CustomerSales
FROM orders
GROUP BY "Customer Name"
)
);"""
df21=pd.read_sql_query(q21,conn)
df21
```

**22. Classify products by profitability**

```python
q22="""SELECT
"Product Name",
SUM(Profit) AS Profit,
CASE
WHEN SUM(Profit) > 500 THEN 'High Profit'
WHEN SUM(Profit) > 100 THEN 'Medium Profit'
ELSE 'Low Profit'
END AS Profit_Category
FROM orders
GROUP BY "Product Name";"""
df22=pd.read_sql_query(q22,conn)
df22
```
